```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd

data = pd.read_csv("books.csv")
data.columns
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data)
plt.show()
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data, hue="Genre")
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data, size=10)
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data, style="Genre")
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data, hue="Genre",palette="Accent")
```

```python
sns.lineplot(x="User Rating", y= "Reviews", data= data, hue= "Genre", markers=["o",">"])
```
